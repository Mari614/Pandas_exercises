### Step 1. Import the necessary libraries

```python
import pandas as pd 
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo=pd.read_csv (url , sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price']=chipo['item_price'].str.replace('$', '')
chipo['item_price']=chipo['item_price'].astype(float) 
```

```python
chipo.dtypes
```

```python
chipo['item_price'] = chipo['item_price']/chipo['quantity']
chipo['quantity'] = 1 #Dividing item_price by quantity, therefore let quantity be 1
chipo.drop_duplicates(['item_name'], keep='first', inplace=True)
chipo.sort_values(by='item_price', ascending=False, inplace=True)
chipo[chipo['item_price']>10]['item_name'].nunique()
#np.count_nonzero(chipo.item_price>10.00)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.groupby(by ="item_name")[["item_price"]].min().reset_index()
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by = "item_name",ascending=True).reset_index(drop=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo[chipo['item_price'] == chipo['item_price'].max()].shape[0]

```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo.item_name == "Veggie Salad Bowl"].quantity.sum()
```
